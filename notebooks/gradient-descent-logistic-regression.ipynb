### Gradient Descent with Logistic Loss Function for Logistic Regression

```python
import numpy as np
import copy
import math
```

```python
### Create mini data set for binary classification
X_train = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]]) # 6 samples, 2 features
y_train = np.array([0, 0, 0, 1, 1, 1])
```

```python
### Create logistic regression function:
def sigmoid(z):
    """Logistic function mapping any value to a value between 0 and 1."""
    return 1 / (1 + np.exp(-z))


def logistic_regression(X, w, b):
    """Logistic regression function that takes in data and weights and returns predicted probabilities."""
    z = np.dot(X, w) + b
    return sigmoid(z)
```

```python
### Compute cost function for logistic regression:
def compute_cost_logistic(X, y, w, b):
    m = X.shape[0]  # number of samples
    cost = 0.0
    for i in range(m):
        f_wb_i = logistic_regression(X[i], w, b) # predicted probability for sample i
        loss = -y[i] * np.log(f_wb_i) - (1 - y[i]) * np.log(1 - f_wb_i) # loss for sample i using cross-entropy loss
        cost += loss
    cost /= m
    return cost
```

```python
### Computation of the gradient for logistic regression:
def compute_gradient_logistic(X, y, w, b):
    """Compute the gradient for logistic regression."""
    m,n = X.shape
    dj_dw = np.zeros(n,)
    dj_db = 0.
    
    for i in range(m):
        f_wb_i = logistic_regression(X[i], w, b) # comute y hat for i-th example
        err_i = f_wb_i - y[i] # compute error for i-th example
        for j in range(n):
            dj_dw[j] = dj_dw[j] + err_i * X[i, j] # compute dj_dw (partial derivative) for j-th feature of the i-th example
        dj_db = dj_db + err_i # compute dj_db (partial derivative) for the i-th example
    
    dj_dw = 1/m * dj_dw # average over all examples
    dj_db = 1/m * dj_db # average over all examples

    return dj_dw, dj_db
```

```python
def gradient_descent_logistic(X, y, w_in, b_in, alpha, num):
    w = copy.deepcopy(w_in)
    b = b_in
    
    J_history = []
    
    for i in range(num):
        dj_dw, dj_db = compute_gradient_logistic(X, y, w, b)
        w = w - alpha * dj_dw
        b = b - alpha * dj_db
    
        if i < 100000: # prevent resource exhaustion
            J_history.append(compute_cost_logistic(X, y, w, b))
        if i % math.ceil(num / 10) == 0:
            print(f"Iteration {i:4d}: Cost {J_history[-1]}   ")
    
    return w, b, J_history
```

```python
w_tmp  = np.zeros_like(X_train[0])
b_tmp  = 0.
alph = 1
iters = 10000

w_out, b_out, _ = gradient_descent_logistic(X_train, y_train, w_tmp, b_tmp, alph, iters) 
print(f"\nupdated parameters: w:{w_out}, b:{b_out}")
```
