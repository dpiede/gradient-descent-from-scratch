### Gradient Descent with Logistic Loss Function for Logistic Regression

```python
import numpy as np
```

```python
### Create mini data set for binary classification
X_train = np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]]) # 6 samples, 2 features
y_train = np.array([0, 0, 0, 1, 1, 1])
```

```python
### Create logistic regression function:
def sigmoid(z):
    """Logistic function mapping any value to a value between 0 and 1."""
    return 1 / (1 + np.exp(-z))


def logistic_regression(X, w, b):
    """Logistic regression function that takes in data and weights and returns predicted probabilities."""
    z = np.dot(X, w) + b
    return sigmoid(z)
```

```python
### Computation of the gradient for logistic regression:
def compute_gradient_logistic(X, y, w, b):
    """Compute the gradient for logistic regression."""
    m,n = X.shape
    dj_dw = np.zeros(n,)
    dj_db = 0.
    
    for i in range(m):
        f_wb_i = logistic_regression(X[i], w, b) # comute y hat for i-th example
        err_i = f_wb_i - y[i] # compute error for i-th example
        for j in range(n):
            dj_dw[j] = dj_dw[j] + err_i * X[i, j] # compute dj_dw for j-th feature of the i-th example
        dj_db = dj_db + err_i # compute dj_db for the i-th example
    
    dj_dw = 1/m * dj_dw # average over all examples
    dj_db = 1/m * dj_db # average over all examples

    return dj_dw, dj_db
```

```python
# testing to retrieve the gradient for logistic regression
compute_gradient_logistic(X_train, y_train, np.array([2.0, 3.0]), 1)
```
