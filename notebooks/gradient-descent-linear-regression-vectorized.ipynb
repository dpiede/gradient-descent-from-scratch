#### Gradient Descent Vectorized Implementation (Multiple Linear Regression)

```python
import numpy as np
```

```python
### Creating a small dataset
X_train = np.array([[2104, 5, 1, 45], [1416, 3, 2, 40], [852, 2, 1, 35]])
y_train = np.array([460, 232, 178])
```

```python
### Since we have now multiple features, we also get a vector of parameter w, instead of a scalar.
b_init = 0.0
w_init = np.zeros(X_train.shape[1]) # initialize w as a vector of zeros
print(f"w_init = {w_init}, b_init = {b_init}")
```

```python
### Now we can compute predictions using the dot product of w and x, instead of multiple multiplications and additions.
def predict(w, b, x):
    return np.dot(w, x) + b # using the dot product for simultaneous multiplication and addition
preditions = predict(w_init, b_init, X_train[0])
print(f"Prediction for the first training example: {preditions}")
```

```python
### Same foes for cost computation, as we can use vectorized operations for all training examples at once.
def compute_cost(w, b, X, y):
    m = X.shape[0]
    cost = 0.0
    f_wb = X @ w + b # using matrix multiplication for all training examples at once
    cost = np.sum((f_wb - y) ** 2) / (2 * m) # vecrtorized cost computation
    return cost
```

```python
computed_cost = compute_cost(w_init, b_init, X_train, y_train)
print(f"Cost at initial w and b: {computed_cost}")
```

```python
def compute_gradient_vectorized(w, b, X, y):
    m = X.shape[0]
    f_wb = X @ w + b # predictions for all training examples
    error = f_wb - y # error vector for all training examples
    dj_dw = (X.T @ error) / m # gradient with respect to w using matrix multiplication
    dj_db = np.sum(error) / m # gradient with respect to b
    return dj_db, dj_dw
```

```python
# Compute gradient descent updates using vectorized operations
def gradient_descent(w, b, X, y, alpha, num_iters):
    J_history = []

    for i in range(num_iters):
        dj_db, dj_dw = compute_gradient_vectorized(w, b, X, y)
        w = w - alpha * dj_dw
        b = b - alpha * dj_db
        if i < 100000: # prevent resource exhaustion
            J_history.append(compute_cost(w, b, X, y))
    return w, b, J_history
```

```python
params = gradient_descent(w_init, b_init, X_train, y_train, alpha=5.0e-7, num_iters=100000)
```

```python
print(f"Updated parameters after gradient descent: w = {params[0].round(2)}, b = {params[1].round(2)}")
print(f"Cost after gradient descent: {params[2][-1].round(2)}")
```
